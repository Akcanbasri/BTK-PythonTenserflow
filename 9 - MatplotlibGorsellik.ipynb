```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
numpyDizisi1 = np.linspace(1,10,20)
```

```python
numpyDizisi2 = numpyDizisi1 ** 2
```

```python
(benimFigure, benimEksen) = plt.subplots()
benimEksen.plot(numpyDizisi1,numpyDizisi2,color="#FF1493", alpha = 0.5) # alpha saydamlık veriyor
benimEksen.plot(numpyDizisi2, numpyDizisi1, color= "#00FFFF")
```

```python
(yeniFigure, yeniEksen) = plt.subplots()
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 2, color="blue",linewidth= 2.0)
yeniEksen.plot(numpyDizisi1,numpyDizisi1 + 3, color = "yellow", linewidth= 2.0)

yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 4, color= "#000000", linestyle= "-.")
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 6, color= "#556B2F", linestyle= "--")
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 7, color= "#DC143C", linestyle= ":")

yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 8, color= "#000000", linestyle= "--", marker = "o", markersize= 8, markerfacecolor="red")
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 9, color="#000000", linestyle="--",marker="+",markersize= 4)
```
