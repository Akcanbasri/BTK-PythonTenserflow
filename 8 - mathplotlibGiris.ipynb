```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
yasListesi = [10,20,30,30,30,40,50,60,70,75]
kiloListesi = [20,60,80,85,86,87,70,90,95,90]

numpyYasDizisi = np.array(yasListesi)
numpyKilolistesi = np.array(kiloListesi)
```

```python
plt.plot(numpyYasDizisi,numpyKilolistesi,"y")
plt.xlabel("Yas")
plt.ylabel("Kilo")
plt.title("Kilonun yasa g√∂re degisim grafigi")
# plt.show()
```

```python
numpyDizisi1 = np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 ** 3
```

```python
numpyDizisi2
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"g*-")
```

```python
plt.subplot(1,2,1)
plt.plot(numpyDizisi1,numpyDizisi2,"g*-")
plt.subplot(1,2,2) # 1 sƒ±ra olacak. 2 kolon olacak ve ≈üuanda ikinci grafiƒüi √ßiziyorum.
plt.plot(numpyDizisi2,numpyDizisi1,"r--")
```

```python
benimFigure = plt.figure()
figureAxes = benimFigure.add_axes([0.1,0.1,0.3,0.3]) # [,,uzunluƒüu,y√ºksekliƒüi] verir
figureAxes.plot(numpyDizisi1,numpyDizisi2,"g")
figureAxes.set_xlabel("X ekseni verisi")
figureAxes.set_ylabel("Y ekseni verisi")
figureAxes.set_title("Grafik ba≈ülƒ±gƒ±")
```

```python
figure2 =plt.figure()
eksen1 = figure2.add_axes([0.1,0.1,0.9,0.9])
eksen2 = figure2.add_axes([0.2,0.6,0.3,0.3])

eksen1.plot(numpyDizisi1,numpyDizisi2,"r")
eksen1.set_xlabel("X ekseni verisi")
eksen1.set_ylabel("Y ekseni verisi")
eksen1.set_title("Eksen bir grafik basligi")

eksen2.plot(numpyDizisi2,numpyDizisi1,"g")
eksen2.set_xlabel("X ekseni verisi")
eksen2.set_ylabel("Y ekseni verisi")
eksen2.set_title("ic grafik basligi")
```

```python
(benimFigure, benimEksenler) = plt.subplots(nrows=1,ncols=2)
type(benimEksenler)

for eksen in benimEksenler:
    eksen.plot(numpyDizisi1,numpyDizisi2,"g")
    eksen.set_xlabel("X Ekseni")

plt.tight_layout()
#benimFigure
```

```python
yeniFigure = plt.figure(figsize=(6,6),dpi=100)

yeniEksen = yeniFigure.add_axes([0.1,0.1,0.9,0.9])
yeniEksen.plot(numpyDizisi1,numpyDizisi1 ** 2 ,label = "numpyDizisi ** 2")
yeniEksen.plot(numpyDizisi1, numpyDizisi1 ** 3,label = "numpyDizisi ** 3")
yeniEksen.legend(loc=6)
```
